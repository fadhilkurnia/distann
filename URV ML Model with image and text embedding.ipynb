A model that will take images or queries and convert them into vector embeddings. I dont want to put in the dataset here, but rather create a function that will take an image or query as input and output the vector embedding of it. Will be using the OpenAI CLIP model and MS-COCO Dataset.


store the vector as binary

```python
!pip install transformers
!pip install torch
import torch
from PIL import Image
from transformers import CLIPModel, CLIPProcessor
# !git clone <https://github.com/fadhilkurnia/distann.git>
# %cd distann
# !git checkout Anand
```

```python
model_name = "openai/clip-vit-base-patch32"
model = CLIPModel.from_pretrained(model_name)
processor = CLIPProcessor.from_pretrained(model_name)
```

```python
def vectorize_image(image_path):
    image = Image.open(image_path)
    inputs = processor(images=image, return_tensors="pt")
    with torch.no_grad():
        image_features = model.get_image_features(**inputs)
    return image_features
```

```python
def vectorize_text(text):
    inputs = processor(text, return_tensors="pt")
    with torch.no_grad():
        text_features = model.get_text_features(**inputs)
    return text_features
```

```python
user_input = input("Enter a query: ")
embedding = vectorize_text(user_input)
```
